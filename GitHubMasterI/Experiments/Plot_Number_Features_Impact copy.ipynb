```python
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
import numpy as np
from matplotlib.lines import Line2D
```

```python
# Create a function to generate smooth curves
def smooth_curve(x, y, points=1000):
    spl = UnivariateSpline(x, y)
    xs = np.linspace(min(x), max(x), points)
    ys = spl(xs)
    return xs, ys
```

```python
# Dataset 1
x1 = [15, 30, 45, 60]
f1_score1 = [0.72, 0.74, 0.75, 0.76]
precision1 = [0.74, 0.75, 0.75, 0.77]
recall1 = [0.73, 0.74, 0.75, 0.76]

# Dataset 2
x2 = [15, 30, 45, 60]
f1_score2 = [0.73, 0.74, 0.77, 0.77]
precision2 = [0.75, 0.75, 0.76, 0.77]
recall2 = [0.74, 0.74, 0.77, 0.77]

# Extract colors from the viridis colormap
colors = plt.cm.viridis(np.linspace(0, 1, 3))

plt.figure(figsize=(9,10))

# Plot for Dataset 1 (curvy solid lines with thicker width)
x_smooth1, f1_smooth1 = smooth_curve(x1, f1_score1)
_, precision_smooth1 = smooth_curve(x1, precision1)
_, recall_smooth1 = smooth_curve(x1, recall1)
plt.plot(x_smooth1, f1_smooth1, color=colors[0], linewidth=5, linestyle='-')  # Increased linewidth
plt.plot(x_smooth1, precision_smooth1, color=colors[1], linewidth=5, linestyle='-')  # Increased linewidth
plt.plot(x_smooth1, recall_smooth1, color=colors[2], linewidth=5, linestyle='-')  # Increased linewidth

# Plot for Dataset 2 (curvy dashed lines)
x_smooth2, f1_smooth2 = smooth_curve(x2, f1_score2)
_, precision_smooth2 = smooth_curve(x2, precision2)
_, recall_smooth2 = smooth_curve(x2, recall2)
plt.plot(x_smooth2, f1_smooth2, color=colors[0], linewidth=2, linestyle='--')
plt.plot(x_smooth2, precision_smooth2, color=colors[1], linewidth=2, linestyle='--')
plt.plot(x_smooth2, recall_smooth2, color=colors[2], linewidth=2, linestyle='--')

# Adding titles and labels
plt.title('XGB: Metrics over number of features in external validation', fontsize=18)
plt.xlabel('Number of features', fontsize=16)
plt.ylabel('Score', fontsize=16)
plt.xticks(x1)

# Display the grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Create custom legend
legend_elements = [Line2D([0], [0], color=color, lw=3, label=label) for color, label in zip(colors, ['F1-score', 'Precision', 'Recall'])]
legend_elements += [Line2D([0], [0], color='black', lw=4.5, linestyle='-', label='External set I'),  # Increased linewidth for legend
                    Line2D([0], [0], color='black', lw=3, linestyle='--', label='External set II')]

plt.legend(handles=legend_elements, fontsize='large')

# Save the figure with high quality
plt.tight_layout()
plt.savefig('XGB_metrics_plot_curvy_thickness.png', dpi=300)

plt.show()
```
