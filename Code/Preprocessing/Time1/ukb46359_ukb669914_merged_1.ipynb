# Merge clean data

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
df_ex = pd.read_csv('/Users/marinacamacho/Desktop/Master_I/Raw_Data/Time_1/ukb46359_clean_external.csv',delimiter="\,").sort_values(by=['f.eid']).reset_index(drop=True)
df_in = pd.read_csv('/Users/marinacamacho/Desktop/Master_I/Raw_Data/Time_1/ukb46359_clean_internal.csv',delimiter="\,").sort_values(by=['f.eid']).reset_index(drop=True)
df_ec = pd.read_csv('/Users/marinacamacho/Desktop/Master_I/Raw_Data/Time_1/ukb46359_clean_earlycause.csv',delimiter="\,").sort_values(by=['f.eid']).reset_index(drop=True)
df_new = pd.read_csv('/Users/marinacamacho/Desktop/Master_I/Raw_Data/Time_1/ukb669914_clean.csv',delimiter="\,").sort_values(by=['f.eid']).reset_index(drop=True)
```

```python
df_ec = df_ec.loc[df_ec['f.eid'].isin(df_new['f.eid'])].sort_values(by=['f.eid']).reset_index(drop=True)
df_ex = df_ex.loc[df_ex['f.eid'].isin(df_new['f.eid'])].sort_values(by=['f.eid']).reset_index(drop=True)
df_in = df_in.loc[df_in['f.eid'].isin(df_new['f.eid'])].sort_values(by=['f.eid']).reset_index(drop=True)
df_new = df_new.loc[df_new['f.eid'].isin(df_ec['f.eid'])].sort_values(by=['f.eid']).reset_index(drop=True)
```

```python
df_ec.shape, df_ex.shape, df_in.shape, df_new.shape
```

```python
del df_ex['f.eid']
del df_in['f.eid']
del df_new['f.eid']
```

```python
df_all = pd.concat([df_ec,df_ex,df_new,df_in],axis=1)
df_ext = pd.concat([df_ec,df_ex,df_new],axis=1)
```
